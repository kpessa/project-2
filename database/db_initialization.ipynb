```python
# Dependancies
import pandas as pd
import os
from sqlalchemy import create_engine

# Import relevant files
# File load
file1 = os.path.join('../static/data/complete_data.csv')
file2 = os.path.join('../static/data/fl_data.csv')

# File read
complete_df = pd.read_csv(file1)
fl_df = pd.read_csv(file2)
```

```python
# Sql lite file name
sql_lite = "sqlite:///covid_factors_database.sqlite"
```

```python
#  Create engine
engine = create_engine(sql_lite)
```

```python
complete_df.to_sql(name="complete_data", con=engine, if_exists="replace", index=False)
fl_df.to_sql(name="florida_data", con=engine, if_exists="replace", index=False)
```

```python
engine.table_names()
```

```python
# Testing
df = pd.read_sql_query('SELECT * FROM complete_data', con=engine)
df
```

```python
# Testing
df = pd.read_sql_query('SELECT * FROM florida_data', con=engine)
df
```

```python
engine = engine.dispose()
```
